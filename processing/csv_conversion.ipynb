```python
image_library = pd.read_csv('../data/train-calibrated-shuffled.csv', index_col = 0)
```

```python
image_library
```

```python
start_time = time.time()
label_list = []
img_list = []
for i, image_entry in image_library.iterrows():
    if i % 100 == 0:
        print(f'Loading image {i} at time {round(time.time() - start_time)} s...')
    is_self = image_entry['label'] in [8, 9]
    with Image.open('../data/processed/' + image_entry['origin']) as image:
        img_list.append(np.array(image.getdata(), np.uint8))
        label_list.append(is_self)

with open('../data/processed_csv/train_images.csv', 'w') as file:
    csv_writer = writer(file)
    
    for i, image in enumerate(img_list):
        if i % 100 == 0:
            print(f'Writing line {i} at time {round(time.time() - start_time)} s...')
        csv_writer.writerow(list(image) + [label_list[i]])

print(f'Done. Took {round(time.time() - start_time)} seconds.')
del label_list
del img_list
```
